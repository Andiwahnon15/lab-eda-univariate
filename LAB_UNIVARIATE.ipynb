```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\andre\Downloads\Murder\LAB\amz_uk_price_prediction_dataset.csv')
```

```python
# PARTE 1
# Generar una tabla de frecuencia para la categor√≠a del producto
frequency_table = df['category'].value_counts()
print(frequency_table.head(5))
```

```python
# Gr√°fico de barras para la distribuci√≥n de productos en diferentes categor√≠as
top_categories = frequency_table.head(10)  # Tomando las 10 categor√≠as m√°s comunes como ejemplo
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.xticks(rotation=45)
plt.title('Distribuci√≥n de Productos en Categor√≠as Principales')
plt.xlabel('Categor√≠a')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gr√°fico circular para las proporciones de las categor√≠as principales
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%')
plt.title('Proporciones de Productos en Categor√≠as Principales')
plt.show()
```

```python
# PARTE 2
# Calcular la media, la mediana y la moda para el precio de los productos
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()

print(f"Media de precios: {mean_price}, Mediana de precios: {median_price}, Moda de precios: {mode_price}")
```

```python
# Calcular la varianza, la desviaci√≥n est√°ndar, el rango y el rango intercuart√≠lico para el precio de los productos
variance_price = df['price'].var()
std_dev_price = df['price'].std()
price_range = df['price'].max() - df['price'].min()
price_iqr = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Varianza de precios: {variance_price}, Desviaci√≥n est√°ndar de precios: {std_dev_price}, Rango de precios: {price_range}, Rango intercuart√≠lico de precios: {price_iqr}")
```

```python
# Histograma para la distribuci√≥n de precios de productos
plt.hist(df['price'].head(20), bins='auto')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Precios de Productos')
plt.show()
```

```python
# Box plot para visualizar la distribuci√≥n de precios de productos
plt.boxplot(df['price'])
plt.ylabel('Precio')
plt.title('Boxplot de Precios de Productos')
plt.show()
```

```python
# PARTE 3 
# Calcular la media, la mediana y la moda para la calificaci√≥n de los productos
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()
print(f"Media de calificaciones: {mean_rating}, Mediana de calificaciones: {median_rating}, Moda de calificaciones: {mode_rating}")
```

```python
# Calcular la varianza, la desviaci√≥n est√°ndar y el rango intercuart√≠lico para las calificaciones de los productos
variance_rating = df['stars'].var()
std_dev_rating = df['stars'].std()
iqr_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)
print(f"Varianza de calificaciones: {variance_rating}, Desviaci√≥n est√°ndar de calificaciones: {std_dev_rating}, Rango intercuart√≠lico de calificaciones: {iqr_rating}")
```

```python
# Calcular el sesgo y la curtosis para las calificaciones de los productos
skewness_rating = df['stars'].skew()
kurtosis_rating = df['stars'].kurt()
print(f"Sesgo de calificaciones: {skewness_rating}, Curtosis de calificaciones: {kurtosis_rating}")
```

```python
# Histograma para la distribuci√≥n de calificaciones de productos
plt.hist(df['stars'], bins='auto')
plt.xlabel('Calificaci√≥n')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de Calificaciones de Productos')
plt.show()
```
